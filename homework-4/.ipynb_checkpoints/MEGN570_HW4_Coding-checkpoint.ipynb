# MEGN 570 HW 4 - Charge Transfer Modeling

We will model charge transfer at an SOFC anode, where O$^{2-}$ in the electrolyte reacts with 2 hydrogen atoms adsorbed on a Ni surface (H$_{\rm Ni(s)}$) to form adsorbed water (H$_2$O$_{\rm Ni(s)}$) and electrons in the Ni (e$^-_{\rm Ni}$).

$$\mathrm{O^{2-}_{elyte(s)} + 2\,H_{Ni(s)} \leftrightarrows \left[\,\right]_{elyte(s)} + H_2O_{Ni(s)} + \left[\,\right]_{Ni(s)}+ 2\,e^-_{Ni}  }$$

where $\left[\,\right]_{\rm elyte}$ is a surface vacancy on the ceramic electrolyte surface, where the oxide formerly resided, and $ \left[\,\right]_{Ni(s)}$ is a vacancy on the Ni surface.  

We are going to plot the charge transfer current density as a function of the double layer potential $\Delta\phi = \phi_{\rm an} - \phi_{\rm elyte}$ for $-1.0 \leq \Delta \phi \leq 0.15\,{\rm V}$.

For this reaction, the standard concentration for the Ni surface is the total site density, $C^\circ_{\rm Ni(s)} = \Gamma_{\rm Ni(s)} = 2.6\times10^{-8}\,\frac{\rm kmol}{m^2}$, and for the electrolyte it is the total electrolyte surface site density: $C_{\rm elyte}^\circ =\Gamma_{\rm YSZ(s)} =  46.05\,\frac{\rm kmol}{\rm m^2}$

The species thermo and forward rate coefficient are provided for you.

## Part 1:  Elementary mass action kinetics.

Calculate the current density as a function of the electric potential, using elementary mass-action kinetics.  The only requirement is that you provide an equation for `i_elementary[i]`, which calculates the current density ($A m^{-2}$) for each electric potential value.  How you get to that value is up to you.

The actual answers are plotted with your equation, to verify.  

Note that we define positive current as positive charge moving _to_ the electrode, and the electric potential as $\Delta \phi = \phi_{\rm an} - \phi_{\rm elyte}$.  As we are sweeping from $\Delta \phi=0.15$ to $\Delta \phi = -1.0$, should the current get more positive or more negative?

Assume the following conditions:

- $\theta_{\rm H_{Ni(s)}} = 0.6$
- $\theta_{\rm H_2O_{Ni(s)}} = 0.2$
- $\theta_{\rm \left[\,\right]_{Ni(s)}} = 0.2$
- $\theta_{\rm O^{2-}_{elyte(s)}} = 0.92$
- $\theta_{\rm \left[\,\right]_{elyte(s)}} = 0.08$
- $T = 800^\circ$ C

The $k_{\rm fwd}$ value given below is in units that will results in a current given in A m$^{-2}$.

### Import modules:

```python
import numpy as np
from matplotlib import pyplot as plt
from math import exp, log
```

### Givens

```python
F = 96485e3    # Faraday's constant, C/kmol equivalent charge
R = 8314.5     # Gas constant, J/kmol-K

k_fwd = 4.16307062e+7 # Chemical forward rate constant, m^4/kmol^2/s

C_elyte = 46.05    # Total (reference) elyte concentration, kmol/m3
C_Ni_s = 2.6e-08   # Total (reference) concentration of Ni surface sites, kmol/m2

X_H_Ni = 0.6
X_H2O_Ni = 0.2
X_Vac_Ni = 0.2
X_Vac_elyte = 0.08
X_Ox_elyte = 0.92

"Species standard-state thermo"
g_H_Ni_o = -7.109209e+07      # standard-state gibbs energy for H adsorbed on Ni surface (J/kmol)
g_H2O_Ni_o = -3.97403035e+08  # standard-state gibbs energy for H2O adsorbed on Ni surface (J/kmol)
g_Vac_Ni_o = 0.0              # standard-state gibbs energy for Ni surface vacancy (J/kmol)
g_Vac_elyte_o = 0.0           # standard-state gibbs energy for electrolyte oxide vacancy (J/kmol)
g_Ox_elyte_o = -2.1392135e+08 # standard-state gibbs energy for electrolyte oxide O2- (J/kmol)

# Validated (i.e. correct :) results.  We will use these to check our answers:
E_validate = np.array([-0.9, -0.86724138, -0.83448276, -0.80172414, -0.76896552, 
                    -0.7362069, -0.70344828, -0.67068966, -0.63793103, -0.60517241, 
                    -0.57241379, -0.53965517, -0.50689655, -0.47413793, -0.44137931, 
                    -0.40862069, -0.37586207, -0.34310345, -0.31034483, -0.27758621, 
                    -0.24482759, -0.21206897, -0.17931034, -0.14655172,  -0.1137931, 
                    -0.08103448, -0.04827586, -0.01551724,  0.01724138,  0.05])
i_validate = np.array([ 131.74413417,   92.44282478,   64.86313049,   45.50798636,   31.9231884,
                    22.38620407,   15.68775772,   10.97849611,    7.66129177,    5.31552803,
                    3.64376548,    2.43402707,    1.53291973,    0.82618438,    0.22420799,
                    -0.34933914,   -0.96718197,   -1.70766206,   -2.66467121,   -3.95955677,
                    -5.75650834,   -8.28337676,  -11.86056524,  -16.94165607,  -24.17092413,
                    -34.46503005,  -49.12925142,  -70.02299007,  -99.79554155, -142.22202185])
```

### Add some calculations to the following code block:

```python
n = 2
T = 800 + 273    #K
Beta = 0.5

# Loop over these delta phi = phi_anode - phi_elyte values:
delta_phi = np.linspace(-0.9,0.05,100)
i_elementary = np.zeros_like(delta_phi)
overpotential = np.zeros_like(delta_phi)

C_Ox_elyte = X_Ox_elyte*C_elyte   #kmol/m3
C_Vac_elyte = X_Vac_elyte*C_elyte   #kmol/m3
C_H_Ni = X_H_Ni*C_Ni_s   #kmol/m2
C_Vac_Ni = X_Vac_Ni*C_Ni_s   #kmol/m2
C_H2O_Ni = X_H2O_Ni*C_Ni_s   #kmol/m2

prod_C_fwd = C_Ox_elyte**(1)*C_H_Ni**(2)
prod_C_rev = C_Vac_elyte**(1)*C_H2O_Ni**(1)*C_Vac_Ni**(1)

g_rxn_o = g_Vac_elyte_o + g_H2O_Ni_o + g_Vac_Ni_o - g_Ox_elyte_o - 2*g_H_Ni_o    #Delta G_o of reaction (J/kmol)
g_rxn = g_rxn_o + R*T*log(prod_C_rev/prod_C_fwd)
print('g_rxn =', g_rxn)

k_rev = k_fwd/exp((-1)*g_rxn/R/T)*prod_C_fwd/prod_C_rev   #reverse rate constant (m^4/kmol^2/s)

n_charge = -2    #value of electric charge transferred (O 2-)

delta_phi_eq = g_rxn/n/F    #negative sign results in phi_elyte - phi_anode
print('delta_phi_eq =', delta_phi_eq)

for ind, E in enumerate(delta_phi):
    
    k_fwd_phi = k_fwd*exp(-Beta*n_charge*F*delta_phi[ind]/R/T)
    k_rev_phi = k_rev*exp((1-Beta)*n_charge*F*delta_phi[ind]/R/T)
    
    i_elementary[ind] = n_charge*F*(k_fwd_phi*prod_C_fwd - k_rev_phi*prod_C_rev)
    
    overpotential[ind] = -delta_phi[ind] + delta_phi_eq

    
```

Plot the results:

```python
fig1, ax1 = plt.subplots()
ax1.plot(delta_phi,i_elementary,linewidth = 1.5,color = 'k', label='Mass Action')
ax1.plot(E_validate,i_validate,'ro',markerfacecolor='none', label='Data')
ax1.set_xlabel('Electric Potential [V]',family='Times New Roman',fontsize=14)
ax1.set_ylabel('Current [A/m2]',family='Times New Roman',fontsize=14)
plt.legend(frameon=False, loc=9)
ax2 = ax1.twinx()
ax2.plot(delta_phi, overpotential, '--', label='Overpotential')
ax2.set_ylabel('Overpotential [V]',family='Times New Roman',fontsize=14)
plt.legend(frameon=False, loc=8)
```

```python
fig2, ax2 = plt.subplots()
ax2.semilogy(delta_phi,abs(i_elementary),linewidth = 1.5,color = 'k', label= 'Mass Action')
ax2.semilogy(E_validate,abs(i_validate),'ro',markerfacecolor='none', label = 'Data')
plt.legend(frameon=False, loc=9)
ax2.set_xlabel('Electric Potential [V]',family='Times New Roman',fontsize=14)
ax2.set_ylabel('Current [A/m2]',family='Times New Roman',fontsize=14)
```

### Discussion:

The dashed blue lign in Fig. 1 represents the overpotential of the electrolyte-anode junction, and shows that it increases as the electric potential difference $\Delta\phi = \Delta\phi_{anode} - \Delta\phi_{elyte}$ increases (becomes more negative) beyond the equilibrium potential ($\Delta\phi_{eq}$). This greater overpotential drives positive current, which is contributed from the forward reaction, while the contribution from the reverse reaction is negligible. At low overpotentials ($\Delta\phi \approx 0$), contributions from both reactions become significant so deviation from Tafel behavior is observed. When $\Delta\phi$ is more positive than $\Delta\phi_{eq}$, the reverse reaction becomes dominant. This is equivalent to running the fuel cell in electrolyzer mode so $H_2$ and $O_2$ are produced from water. Fig.2 shows that the symmetry of the forward and reverse charge transfer reactions. Choosing a symmetry factor other than 0.5 results in an assymetric graph instead - see next plot.



```python
i_beta = np.zeros(shape=[5,100])

for x in range(1,6):
    beta = x/10
    for ind, E in enumerate(delta_phi):
        
        k_fwd_phi = k_fwd*exp(-beta*n_charge*F*delta_phi[ind]/R/T)
        k_rev_phi = k_rev*exp((1-beta)*n_charge*F*delta_phi[ind]/R/T)
    
        i_beta[:,ind] = n_charge*F*(k_fwd_phi*prod_C_fwd - k_rev_phi*prod_C_rev)
        
    plt.semilogy(delta_phi, abs(i_beta[x-1]), label=beta)
plt.legend(frameon=False, title='Beta')
plt.xlabel('Electric Potential [V]',family='Times New Roman',fontsize=14)
plt.ylabel('Current [A/m2]',family='Times New Roman',fontsize=14)
```

## ii. Part 2: Butler-Volmer kinetics  

Now use your derivation to calculate $i_\circ$ as a function of your elementary kinetic parameters, and then calculate the currents for the same range of potentials.

Add some code to the block below:

```python

i_o = n_charge*F*k_fwd**(1-Beta)*k_rev**(Beta)*prod_C_fwd**(1-Beta)*prod_C_rev**(Beta)

i_BV = np.zeros_like(delta_phi)

for ind, E in enumerate(delta_phi):
    
    eta = -delta_phi[ind] + delta_phi_eq
    
    i_BV[ind] = i_o*(exp(-Beta*n*F*eta/R/T)-exp((1-Beta)*n*F*eta/R/T))
```

#### Let's plot the BV and Marcus currents together:

```python
fig3, ax3 = plt.subplots()
ax3.plot(delta_phi,i_elementary,linewidth = 1.5,color = 'k', label='Mass Action');
ax3.plot(delta_phi,i_BV,'ro',markerfacecolor='none', label='BV');
ax3.set_xlabel('Electric Potential [V]',family='Times New Roman',fontsize=14)
ax3.set_ylabel('Current [A/m2]',family='Times New Roman',fontsize=14)
plt.legend(frameon=False, loc=9)
ax4 = ax3.twinx()
ax4.plot(delta_phi, overpotential, '--', label='Overpotential')
ax4.set_ylabel('Overpotential [V]',family='Times New Roman',fontsize=14)
plt.legend(frameon=False, loc=8)
plt.show()
```

```python
fig2, ax2 = plt.subplots()
ax2.semilogy(delta_phi,abs(i_elementary),color='k',linewidth = 1.5)
ax2.semilogy(delta_phi,abs(i_BV),'ro',markerfacecolor='none')
ax2.set_xlabel('Electric Potential [V]',family='Times New Roman',fontsize=14)
ax2.set_ylabel('Current [A/m2]',family='Times New Roman',fontsize=14)
plt.legend(['Mass Action','Butler Volmer'],frameon=False)
```

#### Discussion:

The Mass Action Kinetic and Butler-Vollmer model are equivalent because B-V is derived from kinetics, so the plots are identical. The use of a constant exchange current density implies constant reaction rate constants and concentrations. The reaction rate constants may vary with time due to degradation of the interface over the fuel cell's lifetime or due to poisoning of the catalyst surface. Rate constants also strongly depend on temperature, so they can vary along spatial coordiantes in the fuel cell if strong temperature gradients exist. The reactant/product concentrations will also vary down the length of the fuel cell as extent of reaction increases.

### C. Tafel current

We can re-use the same parameters ($i_\circ$, $\beta$, etc.) from the Butler-Volmer expression to calculate the current using the Tafel approximation:

```python
i_Tafel = np.zeros_like(delta_phi)

for ind, E in enumerate(delta_phi):
    
    eta = delta_phi[ind] - delta_phi_eq
    
    i_Tafel[ind] = -i_o*exp(-Beta*n*F*eta/R/T)
```

```python
plt.plot(delta_phi,i_elementary,linewidth = 1.5,color = 'k');
plt.plot(delta_phi,i_BV,'o',markeredgecolor='r',markerfacecolor='none');
plt.plot(delta_phi,i_Tafel,'x',markeredgecolor='b',markerfacecolor='none');
plt.xlabel('Electric Potential [V]',family='Times New Roman',fontsize=14)
plt.ylabel('Current [A/m2]',family='Times New Roman',fontsize=14)
plt.legend(['Mass Action','Butler Volmer','Tafel'],frameon=False)
plt.show()
```

### Disucssion: 

While the B-V and M-A models yield identical results, the Tafel approximation only fits one side of the curve because it neglects either the forward or reverse reaction (reverse in this case). This manifests in an absense of current for negative overpotentials where current from the reverse reaction dominates. The Tafel approximation is only valid for small overpotentials because both forward and reverse reaction currents are significant at small overpotentials. This is demonstrated in the next figure where the Tafel curve deviates from B-V and M-A curves as $\Delta\phi$ approaches $\Delta\phi_{eq} = -0.423 V$.

```python
plt.plot(delta_phi,i_elementary,linewidth = 1.5,color = 'k');
plt.plot(delta_phi,i_BV,'o',markeredgecolor='r',markerfacecolor='none');
plt.plot(delta_phi,i_Tafel,'x',markeredgecolor='b',markerfacecolor='none');
plt.xlabel('Electric Potential [V]',family='Times New Roman',fontsize=14)
plt.ylabel('Current [A/m2]',family='Times New Roman',fontsize=14)
plt.legend(['Mass Action','Butler Volmer','Tafel'],frameon=False)
plt.xlim((-0.6,-0.1))
plt.ylim((-5,5))
plt.show()
```

```python
residual = np.zeros_like(delta_phi)

for ind, E in enumerate(delta_phi):
    
    residual[ind] = i_Tafel[ind] - i_BV[ind]
```

```python
plt.plot(overpotential, residual)
plt.xlim(0.5,-0.1)
plt.ylim(-1,2)
plt.xlabel('Overpotential [V]',family='Times New Roman',fontsize=14)
plt.ylabel('Error (Tafel-BV) [A/m2]',family='Times New Roman',fontsize=14)
```

#### Discussion:

The preceding figure shows that the Tafel results in negligible error (deviation from B-V model) at larger overpotentials (more negative $\Delta\phi$), but explodes as the overpotential approaches zero and becomes negative.
